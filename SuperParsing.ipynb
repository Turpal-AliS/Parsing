```python
import requests

url = "https://otziv-otziv.ru/katalog/saundbaryi/samsung-soundbars/samsung-hw-m450-otzyvy.html"

request = requests.get(url)
```

```python
print(request)
```

```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup(request.text, 'lxml')
```

```python
productName = bs.find('h1')
print(productName.text)
print(productName.get_text()) #получить текст из тега h1
```

```python
reviews = bs.find_all('div', 'container-reviews collapsible collapsed')

for review in reviews:
    print("")
    print(review.find('span', 'user').get_text())
    print(review.find('span', 'date').get_text())
    print(review.find('div', 'stars-container').get('title')[-1])
    print(review.get_text().strip().replace('\n', ''))
```

```python
# Находим элемент, содержащий всех производителей
aside = bs.find('aside', {'class': 'sections-category manufacrurer'})
# Находим все ссылки на производителей и выводим их на экран
manufacturersWrapp = aside.find('div', 'content')
manufacturersList = manufacturersWrapp.find('ul')
manufacturers = manufacturersList.find_all('a')
for m in manufacturers:
    print(m.text)
```
